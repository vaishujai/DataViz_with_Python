## Investigating Bias in Movie Ratings

This dataset is complied from <b>FiveThirtyEight</b> dataset from 2015 consisting of 147 films.

It consists of data from reviwes from different platform such as `Metacritic`, `Rotten Tomatoes`,` IMDB` and `Fandango` but our interest lies on Fandango because they also sells tickets from their platform.The data set is available [here](https://github.com/fivethirtyeight/data/tree/master/fandango)

    
Metacritic and Rotten Tomatoes aggregate scores from both users and film critics, while IMDB and Fandango aggregate only from their users. We'll focus on just the average scores from users, because not all of the sites have scores from critics.
<i><u>(This mini project is currently work in progress)</i></u>

```python
import pandas as pd

reviews = pd.read_csv('fandango_scores.csv')
reviews.head()
```

The `RT_user_norm`, `Metacritic_user_nom`, `IMDB_norm`, and `Fandango_Ratingvalue` columns contain the average user rating for each movie, normalized to a 0 to 5 point scale. This allows us to compare how the users on each site rated a movie. While using averages isn't perfect because films with a few reviews can skew the average rating, FiveThirtyEight <b>only selected movies with a non-trivial number of ratings</b> to ensure films with only a handful of reviews aren't included.
